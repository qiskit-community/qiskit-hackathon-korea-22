### Step 1: Protein Folding as a Quantum Problem

양자 문제로서의 단백질 접힘

#### 1-1: Why is Protein Folding a Problem? 왜 단백질 접힘이 어려운 문제인가?

The prediction of the three-dimensional structure of protein from amino acid sequence is important for both scientific community and industries. Let's think about it. Most of the important biological phenomena is related to protein. Note that the "Central Dogma" of biology is a process of synthesizing protein in a cell. From the practical point of view, biotechnology and medicine industries will make a quantum leap when it becomes able to design a sequence of amino acid that gives us the desired 3D structure when folded.  
If everybody recognizes the protein folding problem and its inverse problem, why don't scientists start working on it? In fact, they tried, and tried very hard. But still, they are only able to predict rather folding of short sequences.  
Why is predicting protein folding so hard? The same question we have was asked in 1968 by Cyrus Levinthal. [1] Now called "Levinthal's paradox", his argument goes like this. Consider a polypeptide made of 100 amino acids, resulting in 99 polypeptide bonds. Each bond has two rotational degree of freedom in 3D space. Even when assuming each angle can take one of three stable conformations, the resulting conformation space (or parameter space) has $3^{2\cdot99}=3^{198}$ possible conformations. When assuming that the polypeptide "searches" for the energy ground state in the conformation space within 1 ns, the time of the sequence folding into its ground state should take more then the age of the universe! In reality, the proteins fold within a millisecond scale.  
Although the theoretical answer is given to solve the Levinthal's paradox, it does not mean that humans can solve the protein folding problem in such short time scale: We still have to search for the energy ground state to solve the problem. Nowadays, scientists are trying to solve the problem using the Monte Carlo method. (For biology majors: when approaching the problem in a template-free manner.) Even applying the most advanced tricks to overcome the vastness of the conformation space lead to solutions of relatively small-sized polypeptides. [2]

아미노산 서열로부터 단백질의 3차원 구조를 예측하는 것은 과학계와 산업계 모두에서 중요합니다. 생각해 봅시다. 중요한 생물학적 현상의 대부분은 단백질과 관련이 있습니다. 생물학의 중심원리는 세포에서 단백질이 합성되는 것입니다. 실용적인 관점에서 생명 공학 및 의학 산업은 접었을 때 원하는 3D 구조를 제공하는 아미노산 서열을 설계할 수 있게 되면 비약적인 도약을 할 것입니다.
모든 사람들이 단백질 접힘 문제와 그 반대인 단백짐 풀림이 중요한 문제임을 인식하고 있다면 과학자들이 연구를 시작하지 않는 이유는 무엇입니까? 사실, 과학자들은 매우 열심히 노력했습니다. 그러나 여전히 짧은 시퀀스의 접힘만 예측할 수 있습니다.
단백질 접힘을 예측하는 것이 왜 그렇게 어려운가요? Cyrus Levinthal은 1968년에 동일한 질문을 했습니다. [1] 이제 "레빈탈의 역설"이라고 불리는 그의 주장은 이렇습니다. 100개의 아미노산으로 구성된 폴리펩타이드가 99개의 폴리펩타이드 결합을 생성한다고 생각해 보십시오. 각 결합은 3D 공간에서 2개의 회전 자유도를 갖습니다. 각 각도가 세 가지 안정적인 구조 중 하나를 취할 수 있다고 가정하더라도 결과 구조 공간(또는 매개변수 공간)은 $3^{2\cdot99}=3^{198}$ 가능한 구조를 갖습니다. 폴리펩타이드가 1ns 이내의 형태 공간에서 에너지 바닥 상태를 "검색"한다고 가정할 때, 염기서열이 바닥 상태로 접히는 시간은 우주의 나이보다 더 오래 걸릴 것입니다! 하지만 실제 단백질은 밀리초 단위로 접힙니다.
레빈탈의 역설을 풀기 위한 이론적인 답이 주어졌지만, 인간이 단백질 접힘 문제를 그렇게 짧은 시간에 풀 수 있다는 의미는 아닙니다. 우리는 여전히 문제를 해결하기 위해 에너지 바닥 상태를 찾아야 합니다. 최근 과학자들은 몬테카를로 방법을 사용하여 문제를 해결하려고 하고있습니다. 하지만 가장 진보된 기술을 적용하더라도 비교적 작은 크기의 폴리펩티드의 해법만을 찾을 수 있습니다. [2]

#### 1-2: Basics of Protein Folding 단백질 접힘의 기본원리

One can simply think of the structure of protein as side chains attached to the backbone. Protein backbone is what gives the protein its overall shape, and side chains are the ones sticking out of the protein backbone. There are four hierachies of protein: primary structure, secondary structure, tertiary structure, and quaternary structure. Primary structure is basically an amino acid sequence. When it is folded locally, the secondary structure is formed. Tertiary structure is caused by non-local interactions which are mostly hydrophobic core, but also include hydrogen bonds, disulfide bonds and many others.

단백질의 구조는 뼈대(backbone chain)에 부착된 곁사슬(side chain)로 간단히 생각할 수 있습니다. 단백질의 뼈대는 단백질에 전체적인 모양을 부여하고 곁사슬은 단백질 뼈대에서 튀어나온 것입니다. 단백질에는 1차 구조, 2차 구조, 3차 구조 및 4차 구조의 네가지 계층이 있습니다. 1차 구조는 기본적으로 아미노산 서열입니다. 국부적으로 접히면 2차 구조가 형성됩니다. 3차 구조는 대부분 소수성 코어이지만 수소 결합, 이황화 결합 및 기타 여러 가지를 포함하는 비-국소 상호 작용에 의해 발생합니다.

  
Levels of structure in protein. Image from [3].

As sampling from the entire conformation space is computationally impossible, there have been many approaches to coarse-grain the conformation space. One of the simplest approach is lattice protein model where a polypeptide chain is located on the lattice space. We will take this approach as well as the current stage of quantum computer is still in NISQ (noisy intermediate-scale quantum) era. As given in [5], we choose the tetrahedral lattice.

전체 구조적 공간에서 샘플링하는 것은 계산적으로 불가능하기 때문에 구조 공간을 소산입자화(coarse-grained)하는 방법들을 개발해냈습니다. 가장 간단한 접근법 중 하나는 폴리펩타이드 사슬이 격자 공간에 위치하는 격자 단백질 모델입니다. 이 도전에서는 격자 단백질 모델을 사용할 것이며, NISQ(noisy middle-scale quantum) 시대의 양자컴퓨터를 활용하는 기법을 함께 사용할 것입니다. 특히 [5]에서와 같이 사면체 격자를 사용하겠습니다.

  
Protein on four different types of lattice models. Image from [4]

For more interested readers, I refer you to "Ch 4. The Three-dimensional Structure of Proteins" from "Principles of Biochemistry" by Lehninger. [3]

더 알아보고 싶으신 분께서는 Lehninger의 "Principles of Biochemistry"의 "Ch 4. Three-dimensional Structure of Proteins"를 참조하십시오. [3]

#### 1-3: Modeling Protein 단백질 모델링

##### 1-3-1. Configuration Qubits 큐빗 구성

   
Tetrahedral lattice polymer model. Image from [5].  
As can be seen from the figure above, monomers (or "beads") placed on the tetrahedral lattice model can have two different sublattices. Here, we call them as sublattice $\mathcal{A}$ and sublattice $\mathcal{B}$ where $\mathcal{A}$ and $\mathcal{B}$ differ by allowed growing direction. For $\mathcal{A}$, it is only allowed to grow in $0, 1, 2, 3$ directions, but for $\mathcal{B}$, it is only allowed to grow in $\bar{0}, \bar{1}, \bar{2}, \bar{3}$ directions. In the tetrahedral lattice, sites $\mathcal{A}$ and $\mathcal{B}$ should be alternated, meaning that we can assign sites $\mathcal{A}$ as "even" sites and $\mathcal{B}$ as "odd" sites. 

사면체 격자 폴리머 모델. [5]의 이미지.

위의 그림에서 볼 수 있듯이 사면체 격자 모델에 배치된 monomer(또는 "비드")는 두 가지 다른 하위 격자를 가질 수 있습니다. 여기서 $\mathcal{A}$와 $\mathcal{B}$는 가능한 성장 방향에 따라 달라지는데 이를 부격자 $\mathcal{A}$ 및 부격자 $\mathcal{B}$라고 합니다. $\mathcal{A}$ 의 경우 $0, 1, 2, 3$ 방향으로만 증가할 수 있지만 $\mathcal{A}$ 의 경우 $\bar{0}, \bar{1}, \bar{2}, \bar{3}$ 방향으로만 증가할 수 있습니다. 사면체 격자에서 사이트 $\mathcal{A}$와 $\mathcal{B}$는 교대로 있어야 합니다. 즉, 사이트 $\mathcal{A}$를 "짝수" 사이트로 할당하고 $\mathcal{B}$를 "홀수" 사이트로 지정할 수 있습니다.

How can we encode such information in a quantum computer? As sites are fixed in a tetrahedral lattice, all we have to know is how each sites are connected. For example, if we are looking at a bead placed on site $2i$ with type $\mathcal{B}$, the next bead has to be on site with type $\mathcal{A}$, but we do not know on which site should the bead placed at. Hence, we have to specify how they are connected: Out of four possible growth directions $\bar{0}, \bar{1}, \bar{2}, \bar{3}$, in which direction does the bead on site $2i$ grow to reach the next site of type $\mathcal{A}$?  
Such information can be encoded in quantum computer by encoding turns using one qubit per axis: $t_i=q_{4i-3}q_{4i-2}q_{4i-1}q_{4i}$. We call these qubits used for encoding configuration information as "configuration qubits". Note that if we have a polypeptide sequence of length N, the total number of configuration qubits has to be $N_{cf}=4(N-3)$.

양자 컴퓨터에 이런 정보를 어떻게 인코딩할 수 있을까요? 사이트는 사면체 격자로 고정되어 있으므로 각 사이트가 어떻게 연결되어 있는지만 알면 됩니다. 예를 들어, 유형 $\mathcal{B}$ 인 사이트 $2i$에 배치된 비드가 있는 경우 다음 비드는 유형 $\mathcal{A}$ 인 사이트에 있어야 하지만 비드를 배치해야 하는 사이트를 알지 못합니다. 따라서 우리는 그것들이 어떻게 연결되어 있는지 지정해야 합니다. 4개의 가능한 성장 방향 중 $\bar{0}, \bar{1}, \bar{2}, \bar{3}$, 어느 방향으로 사이트 $2i$의 비드가 성장하여 $\mathcal{A}$ 유형의 다음 사이트에 도달할 수 있을까요?
이러한 정보는 축당 하나의 큐비트를 사용하여 회전을 인코딩하여 양자 컴퓨터에서 인코딩할 수 있습니다: $t_i=q_{4i-3}q_{4i-2}q_{4i-1}q_{4i}$. 구성 정보를 인코딩하는 데 사용되는 이러한 큐비트를 "구성 큐비트"라고 합니다. 길이가 N인 폴리펩타이드 서열이 있는 경우 구성 큐비트의 총 수는 $N_{cf}=4(N-3)$입니다.

##### 1-3-2. Interaction Qubits

Now that we know how to encode information about configuration, we need to now how to encode information about interactions between them. Remember, it is interaction that determines the folding of protein. We introduce new type of qubits called "interaction qubits", which are denoted as $q_{i,n}^{(l)}$ for each $l$th nearest-neighbor interaction on the lattice. For example, $l=1$ interaction describes interaction between beads that are closest to each other, while $l=2$ interaction describes interaction between beads that are next-to-closest to each other. For each $l$th interaction, we assign the pairwise interaction energies $\epsilon_{i,j}^{(l)}$. In this challenge, we only consider the nearest-neighbor interaction ($l=1$) which are descirbed by Miyazawa and Jernigan (MJ) interaction energies.  

앞서 단백질의 구성에 대한 정보를 인코딩하는 방법을 알았으므로 이제 구성 간의 상호 작용에 대한 정보를 인코딩하는 방법을 살펴보겠습니다. 단백질의 접힘을 결정하는 주요 요소는 상호작용입니다. "상호작용 큐비트"라고 하는 새로운 유형의 큐비트를 사용하겠습니다. 이 큐비트는 격자에서 $l$번째 가장 가까운 이웃 상호작용에 대해 $q_{i,n}^{(l)}$으로 표시됩니다. 예를 들어, $l=1$ 상호작용은 서로 가장 가까운 비드 간의 상호작용을 설명하고, $l=2$ 상호작용은 서로 가장 가까운 비드 간의 상호작용을 설명합니다. 각 $l$번째 상호작용에 대해 쌍별 상호작용 에너지 $\epsilon_{i,j}^{(l)}$를 할당합니다. 이 챌린지에서는 Miyazawa 및 Jernigan(MJ) 상호작용 에너지로 설명되는 최근접이웃 상호작용 ($l=1$)만 고려합니다.

##### 1-3-3. Hamiltonian 해밀토니안

Knowing the information about configuration and type of interactions, we can now track the folding procedure that minimizes the Hamiltonian $H(\textbf{q})=H_{gc}(\textbf{q}_{cf})+H_{ch}(\textbf{q}_{cf})+H_{in}(\textbf{q})$ where $\textbf{q}=(\textbf{q}_{cf}, \textbf{q}_{in})$. $H_{gc}$ describes the energy cost for geometrical constraint that governs the growth of backbone, and $H_{ch}$ describes the chirality constraint that governs the growth of sidechains. More interested readers can consult Ref. [5].

구성 및 상호 작용 유형에 대한 정보를 정의했다면 단백질 접힘 과정이 해밀토니안 $H(\textbf{q})=H_{gc}(\textbf{q}_{cf})+H_{ch}(\textbf{q}_{cf})+H_{in}(\textbf{q})$을 최소화 하는 과정임을 알 수 있습니다(이때, $\textbf{q}=(\textbf{q}_{cf}, \textbf{q}_{in})$) . 이때 $H_{gc}$은 뼈대의 성장을 제어하는 기하학적 제약 조건에 대한 에너지 비용을 설명하고 $H_{ch}$는 곁사슬의 성장을 제어하는 키랄성 제약 조건을 설명합니다. 더 관심 있는 독자는 Ref. [5]를 읽어 보시기 바랍니다.

### Step 2: Variational Quantum Eigensolver 변분원리기반 양자 고유값 탐색기

#### 2-1: What is Variational Quantum Eigensolver? 변분원리기반 양자 고유값 탐색기란?

The variational quantum eigensolver (VQE) is developed to approximate the ground state of a given Hamiltonian, which is often a problem of great interest in quantum chemistry or many-body physics. VQE is especially gaining considerable interest due to its applicability to current NISQ devices.

VQE(Variational Quantum eigensolver)는 주어진 해밀토니안의 기저 상태를 탐색하기 위해 개발되었으며, 종종 양자 화학 또는 다물체 물리학에서 많은 과심을 받고 있는 방법입니다. VQE는 현재 NISQ 장치에 적용할 수 있기 때문에 특히 상당한 관심을 받고 있습니다.

  
Visualization of VQE algorithm. Image from [7].

As name suggests, VQE utilizes variational method to find a set of parameters that well-describe the ground state. It goes like this:  
1. Map Hamiltonian of interest to quantum device.  
2. Generate an initial state with a set of parameters.  
3. Put the initial state to the parameterized quantum circuit.  
4. Perform measurements of the expectation value, leading to the energy of the trial state.  
5. Update the set of parameters using an optimization algorithm of your choice.  
6. Repeat steps 2-4 until you get low enough ground state energy.  

Note that steps 1 and 5 are done in a classical computer and steps 2 to 4 are performed in a quantum computer. Interested readers can consult [6] for further information about VQE.


이름에서 알 수 있듯이 VQE는 변분원리를 사용하여 바닥 상태를 가장 잘 설명하는 매개변수의 집합을 찾습니다. 일반적인 계산 과정은 다음과 같습니다.
1. 관심 있는 해밀토니안을 양자 장치에 매핑합니다.
2. 매개변수 세트로 초기 상태를 생성합니다.
3. 초기 상태를 매개변수화된 양자 회로에 넣습니다.
4. 기대값의 측정을 수행하여 시험 상태의 에너지를 얻습니다.
5. 선택한 최적화 알고리즘을 사용하여 매개변수 세트를 업데이트합니다.
6. 바닥 상태 에너지가 충분히 낮아질 때까지 2-4단계를 반복합니다.

1단계와 5단계는 기존 컴퓨터에서 수행되고 2~4단계는 양자 컴퓨터에서 수행됩니다. VQE에 대한 더 많은 정보는 [6]에서 확인할 수 있습니다.

#### 2-2: Many Options for VQE   VQE를 사용하기 위한 여러가지 옵션

Although implementing VQE in Qiskit is a lot of fun, let's see what we can do to improve the performance of VQE by choosing right hyperparameters. There are many options we can tune, but here let's focus on three hyperparameters:  
1. ansatz: In physics, the word "ansatz" is used for smart guesses for solutions of differential equations. In the context of VQE, ansatz refers to a parameterized circuit that is used as an ansatz.   
2. optimizer: Used by the classical computer, optimizer is used to fine-tune your quantum circuit so that it produces as low energy as possible.  

Qiskit에서 VQE를 구현하는 것은 정말 재미있습니다. 올바른 하이퍼파라미터를 선택하여 VQE의 성능을 향상시키기 선택할 수 있는 많은 옵션이 있습니다. 조정할 수 있는 실제 옵션은 더 많지만 여기서는 세 가지 하이퍼파라미터에 중점을 두겠습니다.
1. ansatz: 물리학에서 "ansatz"라는 단어는 미분 방정식의 해에 대한 가설의 의미로 사용됩니다. VQE의 맥락에서, ansatz는 ansatz로 사용되는 매개변수화된 회로를 나타냅니다.
2. optimizer: 클래식 컴퓨터에서 사용하는 옵티마이저는 가능한 한 낮은 에너지를 생성하도록 양자 회로를 미세 조정하는 데 사용됩니다.

### Step 3: Solve Protein Folding using VQE  (단백질 접힘 문제를 VQE로 풀어보기)

#### 3-1: Example System (예제)

```python
from qiskit_nature.problems.sampling.protein_folding.interactions.random_interaction import (
    RandomInteraction,
)
from qiskit_nature.problems.sampling.protein_folding.interactions.miyazawa_jernigan_interaction import (
    MiyazawaJerniganInteraction,
)
from qiskit_nature.problems.sampling.protein_folding.peptide.peptide import Peptide
from qiskit_nature.problems.sampling.protein_folding.protein_folding_problem import (
    ProteinFoldingProblem,
)

from qiskit_nature.problems.sampling.protein_folding.penalty_parameters import PenaltyParameters

from qiskit.utils import algorithm_globals, QuantumInstance

algorithm_globals.random_seed = 23
```

Backbone

뼈대의 정의

```python
main_chain = "APRLRFY"
```

Sidechain

곁사슬의 정의

```python
side_chains = [""] * 7
```

Interaction between Amino Acids

아미노산 간의 상호작용 정의

```python
random_interaction = RandomInteraction()
mj_interaction = MiyazawaJerniganInteraction()
```

Physical Constraints

물리적 제약조건

```python
penalty_back = 10
penalty_chiral = 10
penalty_1 = 10

penalty_terms = PenaltyParameters(penalty_chiral, penalty_back, penalty_1)
```

Peptide Definition

펩타이드 정의

```python
peptide = Peptide(main_chain, side_chains)
```

Protein Folding Problem

단백질 접힘 문제 생성 및 큐비트 operator 생성

```python
protein_folding_problem = ProteinFoldingProblem(peptide, mj_interaction, penalty_terms)
qubit_op = protein_folding_problem.qubit_op()
```

VQE 알고리즘 실행

```python
from qiskit.circuit.library import RealAmplitudes
from qiskit.algorithms.optimizers import COBYLA
from qiskit.algorithms import NumPyMinimumEigensolver, VQE
from qiskit.opflow import PauliExpectation, CVaRExpectation
from qiskit import execute, Aer

# set classical optimizer
optimizer = COBYLA(maxiter=50)

# set variational ansatz
ansatz = RealAmplitudes(reps=1)

# set the backend
backend_name = "aer_simulator"
backend = QuantumInstance(
    Aer.get_backend(backend_name),
    shots=8192,
    seed_transpiler=algorithm_globals.random_seed,
    seed_simulator=algorithm_globals.random_seed,
)

counts = []
values = []


def store_intermediate_result(eval_count, parameters, mean, std):
    counts.append(eval_count)
    values.append(mean)


# initialize CVaR_alpha objective with alpha = 0.1
cvar_exp = CVaRExpectation(0.1, PauliExpectation())

# initialize VQE using CVaR
vqe = VQE(
    expectation=cvar_exp,
    optimizer=optimizer,
    ansatz=ansatz,
    quantum_instance=backend,
    callback=store_intermediate_result,
)

result = vqe.compute_minimum_eigenvalue(qubit_op)
print(result)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure()

plt.plot(counts, values)
plt.ylabel("Conformation Energy")
plt.xlabel("VQE Iterations")

fig.add_axes([0.44, 0.51, 0.44, 0.32])

plt.plot(counts[40:], values[40:])
plt.ylabel("Conformation Energy")
plt.xlabel("VQE Iterations")
plt.show()
```
